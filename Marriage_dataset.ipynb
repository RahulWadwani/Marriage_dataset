# 3. Data Cleaning (Pre - processing Data) 

```python
df.isna().sum()
# there is no null values present in the data but stil we need make the data more cleaner 
```

```python
df.duplicated().sum()
# contains no duplicate values 
```

```python
# cleaning the AgeGroup

age = []
for i in df['AgeGroup']:
    age.append(i[1:-1])

df['AgeGroup'] = age
```

```python
# replacing sex to male and female

df['Sex'] = df['Sex'].replace({'Man':'Male','Woman':'Female'})
```

```python
# deleting the unneccessary column from the table 
del df['Sr.No.']
```

# 4. Exploratory Data Analysis 

```python
# total unique values in each column
for i in range(1,len(df.columns)):
    print(f'Total unqiue value in {df.columns[i]} column',df[df.columns[i]].nunique())
```

```python
# making seperate groups/csv file for different countries for country wise analaysis and further proceedings 
countries = df['Country'].unique()
for country in countries:
    count_list = []
    for i in df.values:
        if country == i[1]:
            count_list.append(i)
    new = pd.DataFrame(data = count_list,columns = df.columns)
    new.to_csv(f'country_wise/{country}.csv',index = False)
```

```python
# feature engineeing 
df['Age'] = df['AgeGroup'].str.extract(r'(\d+)-\d+')
df['Age'] = df['Age'].fillna(0).astype(int)
```

```python
# modeling 
# from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Country'] = le.fit_transform(df['Country'])
df['Target'] = df['MaritalStatus'].apply(lambda x:1 if x == 'Divorced' else 0)
x = df[['Country','Age']]
y = df['Target']
```

```python
# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)

# model training 
model = RandomForestClassifier(random_state = 42)
model.fit(x_train, y_train)

# Prediction
y_pred = model.predict(x_test)

# model validation
# evaluation
print(classification_report(y_test,y_pred,zero_division = 0))
```

```python
# which country has the maximum,minimum number of divorce cases
```

```python
dct = {}
for i in df.values:
    if (i[3] == 'Divorced'):
        if (i[0]) in dct:
            dct[i[0]] += 1
        else:
            dct[i[0]] = 1

country_wise = pd.DataFrame()
country_wise['country_code'] = dct.keys()
country_wise['count'] = dct.values()
```

```python
cc_max = country_wise[country_wise['count'] == max(country_wise['count'])]['country_code'].values[0]
print("Maximum Country Code: ", cc_max)
cc_min = country_wise[country_wise['count'] == min(country_wise['count'])]['country_code'].values[0]
print("Minimum Country Code: ", cc_min)
```

```python
# which age group have maximum divorced cases
age_counts = {}
for i in df.values:
    if i[3] == 'Divorced':
        if (i[1] in age_counts):
            age_counts[i[1]] += 1
        else:
            age_counts[i[1]] = 1

age_count = pd.DataFrame()
age_count['age_groups'] = age_counts.keys()
age_count['count'] = age_counts.values()
ag = age_count[age_count['count'] == max(age_count['count'])]['age_groups'].values[0]
print("Age group with maximum divorced cases: ",ag)
```

```python
# age group with maximum marriages 
age_counter = {}
for i in df.values:
    if i[3] == 'Married':
        if (i[1] in age_counter):
            age_counter[i[1]] += 1
        else:
            age_counter[i[1]] = 1

married = pd.DataFrame()
married['Age_group'] = age_counter.keys()
married['counts'] = age_counter.values()
married_group = married[married['counts'] == max(married['counts'])]['Age_group'].values[0]
print("Age Group with maximum Marriages: ",married_group)
```
